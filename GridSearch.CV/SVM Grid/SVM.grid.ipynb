```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('50_Startups.csv')
```

```python
dataset
```

```python
dataset.columns
```

```python
dataset = pd.get_dummies (dataset,drop_first=True)
```

```python
indep = dataset[['R&D Spend','Administration','Marketing Spend','State_Florida','State_New York']]
dep = dataset[['Profit']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(indep,dep,test_size = 0.03, random_state = 0)
```

```python
#Standardisation
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Grid
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid = {'kernel':['rbf','poly','sigmoid','linear'],
'C':[10,100,1000,2000,3000],'gamma':['auto','scale']}
grid = GridSearchCV(SVR(), param_grid, refit = True, verbose = 3,n_jobs=-1)
# fitting the model for grid search
grid.fit(X_train, Y_train)
```

```python
# print best parameter after tuning
#print(grid.best_params_)
```

```python
re=grid.cv_results_
print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```
