```python
#Question: 1

import pandas as pd


data = {'Customer_ID': [101, 102, 103, 101, 104, 102, 101, 105, 102, 103],
        'Purchase_Amount': [200, 150, 180, 220, 300, 200, 100, 400, 250, 300]}

df = pd.DataFrame(data)

total_purchases = df.groupby('Customer_ID')['Purchase_Amount'].sum().reset_index() 

print("Total Purchases per Customer:") 
print("-" * 30)

print(total_purchases) 
print("-" * 50)


top_3_customers = total_purchases.sort_values(by='Purchase_Amount', ascending=False).head(3)


print("Top 3 Frequent Customers:")
print("-" * 25)

print(top_3_customers)
```

```python
#Question 2:

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {'Square_Feet': [1500, 2000, 2500, 3000, 3500],
        'Price': [300000, 400000, 500000, 600000, 700000]}

df = pd.DataFrame(data)


independent = df [["Square_Feet"]]
dependent = df [["Price"]]


plt.scatter(independent, dependent)
plt.xlabel('Square_Feet', fontsize = 20)
plt.ylabel('Price', fontsize = 20)
plt.show()
print("-" * 100)


from sklearn.model_selection import train_test_split
X_train, X_test,Y_train,Y_test = train_test_split(independent, dependent, test_size =0.30, random_state =0)


from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)



test_data = [[1800], [2800]]
predictions = regressor.predict(test_data)

print("Predicted Prices:")
print(predictions)
```

```python
#Question 3:


import pandas as pd


data = {'Category': ['A', 'B', 'C', 'A', 'D', 'B', 'A', 'E', 'B', 'C', 'C', 'A'],
        'Value': [10, 15, 20, 30, 25, 18, 22, 40, 35, 50, 45, 15]}

df = pd.DataFrame(data)

total_value = df.groupby('Category')['Value'].count().reset_index() 

print("Total Occurrences per Category:") 
print("-" * 30)

print(total_value) 
print("-" * 50)


top_3_customers = total_value.sort_values(by='Value', ascending=False).head(3)


print("Top 3 Frequent Categories:")
print("-" * 25)

print(top_3_customers)
```

```python
#Question 5:

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {'Experience': [1, 2, 3, 4, 5], 
        'Salary': [30000, 35000, 40000, 45000, 50000]}

df = pd.DataFrame(data)

Independent = df [["Experience"]]
Dependent = df [["Salary"]]

from sklearn.model_selection import train_test_split
X_train, X_test,Y_train,Y_test = train_test_split(Independent, Dependent, test_size =0.30, random_state =0)

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)

predictions = regressor.predict([[6]])

print(f"Predicted Salary for 6 years of experience: ${predictions[0][0]:.2f}")

```

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {'word_count': [100, 150, 200, 120, 180, 220], 
        'is_spam': ['ham', 'spam', 'spam', 'ham', 'spam', 'spam']}

df = pd.DataFrame(data)

df = pd.get_dummies (df,drop_first=True)

Independent = df [["word_count"]]
Dependent = df [["is_spam_spam"]]


from sklearn.model_selection import train_test_split
X_train, X_test,Y_train,Y_test = train_test_split(Independent, Dependent, test_size =0.30, random_state =0)

from sklearn.linear_model import LogisticRegression
model=LogisticRegression(random_state=0)
model.fit(X_train, Y_train)     

words = int(input("Enter number of words in email: "))

prediction = model.predict([[words]])

result = "Spam" if prediction[0] == 1 else "Not Spam"

print(f"Prediction for email with {words} words: {result}")




```
