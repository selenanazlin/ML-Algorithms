```python
#2.
import pandas as pd

data = {'A': [1, 2, None],  # Missing value
        'B': [4, None, 6]}  # Missing value

df = pd.DataFrame(data)
mean_value = df.mean()
df = df.fillna(mean_value)  # Error: fillna() does not modify in place

#ans: df is assigent to the fillna()

print(df)

```

```python
#8.
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer

# Creating a dataset with missing values
X_train = np.array([[1, 2], [3, np.nan], [5, 6]])
y_train = np.array([10, 20, 30])

imputer = SimpleImputer(strategy ='mean')
X_train = imputer.fit_transform(X_train)

model = LinearRegression()
model.fit(X_train, y_train)  

print(X_train)
print(model.predict([[2, 3]]))

#ans: SimpleImputer is used to fill the missing values 
```
